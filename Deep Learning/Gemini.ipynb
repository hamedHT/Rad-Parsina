## Summarization Pipeline

```python
from transformers import pipeline

MODEL_NAME = "google/pegasus-large"
summarizer = pipeline(
    "summarization",
    model= MODEL_NAME,
    framework="pt"
)


long_text = """
Ordered by Aileron to disarm the humans, Slide's recklessness led her to open fire on them,
 but the humans retreated unharmed. The Autobots managed to corner one of the dissidents,
 and together they travelled north to meet with his suppliers—who turned out to be Frenzy and Rumble,
 who had facilitated the arms trade to make a quick buck.
  Realizing that any punitive actions would only agitate the situation further,
   Aileron opted to cover up the situation entirely, much to the Devisien's frustration. Aileron,
    at the end of her tether, simply screamed at her to shut up. The Autobots,
     minus Jazz but with Frenzy and Rumble in tow, returned to Autobot City shortly afterwards. Primeless, Part 2
"""

summary_list = summarizer(long_text, max_length=90, min_length=10, do_sample=False)

# چاپ نتایج
print("\n=== متن اصلی ===")
print(long_text)
print("\n=== خلاصه‌ی تولید شده توسط LLM ===")
# نتیجه به صورت یک لیست دیکشنری برگردانده می‌شود
print(summary_list[0]['summary_text'])
```
