```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
url = 'https://www.varzesh3.com/'
driver = webdriver.Chrome()
```

```python
driver.get(url)
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

wait = WebDriverWait(driver, 10)
table = wait.until(EC.presence_of_element_located((By.XPATH, "//div[@role='table']")))

rows = driver.find_elements(By.XPATH, "//div[@role='rowgroup'][2]/div[@role='row']")
rows = rows[0:15]

for row in rows:
    rank = row.find_element(By.XPATH, ".//span[@role='cell'][1]").text
    team = row.find_element(By.XPATH, ".//a[@role='cell']").text
    print(f"تیم {team} در رتبه {rank} قرار دارد.")

team_name = "استقلال"
try:
    rank_element = driver.find_element(By.XPATH, f"//div[@role='rowgroup'][2]/div[@role='row'][.//a[contains(text(), '{team_name}')]]//span[@role='cell'][1]")
    print(f"رتبه تیم {team_name}: {rank_element.text}")
except:
    print("تیم مورد نظر یافت نشد.")
```
