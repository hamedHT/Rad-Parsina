```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

url = 'https://pckala.org/'
driver = webdriver.Chrome()
driver.get(url)

search_box = driver.find_element(By.NAME, value="phrase")
search_box.click()
search_box.send_keys('لپ تاپ')
button = driver.find_element(By.CLASS_NAME, value="promagnifier")
button.click()
time.sleep(5)  

def scroll_until_count(target_count=100, max_scrolls=50):
    previous_count = 0
    no_change_attempts = 0
    scroll_attempt = 0
    
    while scroll_attempt < max_scrolls:
        items = driver.find_elements(By.XPATH, "//div[contains(@class, 'item-area') and contains(@class, 'elm_pg_3') and contains(@class, 'instock')]")
        current_count = len(items)
        print(f"تعداد آیتم‌های فعلی: {current_count}")
        
        if current_count >= target_count:
            print(f"به تعداد هدف ({target_count}) رسیدیم.")
            break
        
        if current_count == previous_count:
            no_change_attempts += 1
            if no_change_attempts >= 3:
                print("تعداد آیتم‌ها افزایش نیافت. احتمالاً به انتهای لیست رسیده‌ایم.")
                break
        else:
            no_change_attempts = 0  # reset
        
        previous_count = current_count
        
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  
        scroll_attempt += 1

scroll_until_count(target_count=100)

all_item = driver.find_elements(
    By.XPATH,
    "//div[contains(@class, 'item-area') and contains(@class, 'elm_pg_3') and contains(@class, 'instock')]"
)

data = []
for item in all_item:
    try:
        name_elem = item.find_element(By.XPATH, ".//h3[@class='product-name']/a")
        name = name_elem.get_attribute("title")
        
        main_price_elem = item.find_element(By.XPATH, ".//del")
        main_price = main_price_elem.text
        main_price = main_price.split('\n')[0]
        
        price_discount_elem = item.find_element(By.XPATH, ".//ins")
        price_discount = price_discount_elem.text
        
        discount_elem = item.find_element(By.XPATH, ".//span[@class='discount-archive-price']")
        discount = discount_elem.text
        
        data.append([name, main_price, price_discount, discount])
        
    except Exception as e:
        print(f"خطا در پردازش یک آیتم: {e}")
        continue

for row in data:
    print(row)

df = pd.DataFrame(data, columns=['name', 'main_price', 'price_discount', 'discount'])
print(f"تعداد کل آیتم‌های استخراج‌شده: {len(df)}")
df.head(50)  
```
