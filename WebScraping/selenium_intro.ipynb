```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
url = 'http://menu.shahrzad-restaurant.com/'
driver = webdriver.Chrome()
```

```python
driver.get(url)
```

```python
categories = driver.find_elements(By.CLASS_NAME, value='nav-tabs-default')
categories
```

```python
for i in categories:
    print(i.text)
```

```python
phone = driver.find_element(By.ID, value='ContentPlaceHolder1_dvwComment_txtContactName')
phone.get_attribute('placeholder')
```

```python
button = driver.find_element(By.NAME, value='ctl00$ContentPlaceHolder1$dvwComment$ctl01')
button.click()
```
